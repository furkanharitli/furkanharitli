```python
numpydiz=np.linspace(10,20,10)
numpydiz2=numpydiz ** 3
```

# "b--"koyarsak grafik çizgi çizgi , "g-*"koyarsak tam noktalarında nokta koyulur,"g*-" yaparsak rastgele noktalar koyulur



```python
plt.plot(numpydiz,numpydiz2,"b-*")
```

```python
plt.plot(numpydiz,numpydiz2,"b--")
```

```python
plt.plot(numpydiz,numpydiz2,"b*-")
```

# plt.subplot( 1= kaç satır grafik olucagını
# plt.subplot( 1, 2= kaç tane grafik olucagını
# plt.subplot( 1, 2 ,1= kaçıncı grafiği çizeceğimizi
#                   gösterirler

```python
plt.subplot(1,2,1)
plt.plot(numpydiz,numpydiz2,"b*-")
plt.subplot( 1, 2 ,2)
plt.plot(numpydiz,numpydiz2,"b*-")
```

# figür bize boş bir grafik alanı sağlar plt.figure()
# figürlerin işe yaraması için exen vermemiz gerekir .add_axes([0.1,0.1,3,1]) 1. boşluk x ekseninin nereden başlayacağını ikinci boşluk y ekseninin nerden başlayacağını 3. boşluk x in uzunluğunu 4. boşluk y nin uzunluğunu verir



```python
benimfigür=plt.figure()
figax=benimfigür.add_axes([0.1,0.1,3,1])
figax.plot(numpydiz,numpydiz2,"b")
```

# grafikler iç içe yazılabilir


```python
figure1=plt.figure()
exen1=figure1.add_axes([0.1,0.1,0.7,0.7])
exen2=figure1.add_axes([0.1,0.3,0.3,0.3])
exen1.plot(numpydiz,numpydiz2)
exen2.plot(numpydiz2,numpydiz)
```

# plt.subplot(1,2) 
# satır 2 sütunu olacak kadar grafik oluşturur yani 2 tane grafik verir eğer numpy dizisi ile çalışırken for döngüsüne sokulmazsa hata verir

```python
(figure1,exen1)=plt.subplots(1,2)
for exen2 in exen1:
    exen2.plot(numpydiz,numpydiz2)
```

# .legend() ve label

# legend ve label ile verileri açıklayarak iç içe grafikler oluşturabiliriz

```python
figure1=plt.figure(dpi=50)
yeniexen=figure1.add_axes([0.1,0.1,0.9,0.9])
yeniexen.plot(numpydiz,numpydiz**2,label="numdiz**2")
yeniexen.plot(numpydiz,numpydiz**3,label="numdiz**3")
yeniexen.legend()
```

# .figure() fonksiyonunun içine 
# figsize=(2,2) yazarsanız grafik kare şeklini alır

# .figure() in içine dpi=100 yazarsanız grafiğiniz 100 piksel olur

```python
figure1.savefig("figür.png",dpi=200)
```
