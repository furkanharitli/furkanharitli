# Lineer Tahminlerde Bulundurtma
* İlk olarak tahmini yaptırtacak modülü çağrıyoruz
* Sonra X ve Y yi belirtiyoruz ki x e göre y yi tahmin ettiricez
* --- fit çalışttırtmayı sağlıyor ---
* ###. dir(model) dediğimizde yapabileceğimiz operasyonları gösterir .###

```python
from sklearn import linear_model
X=df[["TV"]]
Y=df[["sales"]]
reg=LinearRegression()
model=reg.fit(X,Y)
dir(model)
```

* dirdeki metodlar için örnek çalışma:

```python
model.intercept_ #sabit aldırtır ax+c deki c dir
```

```python
model.coef_ #beta 1 katsayısı ax+c deki x dir
```

```python

model.score(X,Y) 
#Yaptığımız model % 60 verileri açıklıyor 
```

* Bu sayfada tv ve satış arasındaki ilişkiyi öğrendik denklem= 7.03+ TV*0,05 dir
* Mesela TV= 1 ise satış=7,08 dir
* Yaptığımız model verilerin %60 nı biliyor
